```python
# Firstly, We need to define the basic functions of operations: addition, subtraction, multiplication, and division.
# These are python function.

def add(x, y):
    return x + y 
# This function adds two numbers x and y and return them using the return statement.
# We have to def more functions for substraction, division, and multiplication.

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y): 
     if y==0:
        return "Error, Division by zero."
     else:
        return x / y 
# Notice that the divide() function is litle bit different. 
# This is because dividing by zero causes an error, so we return a custom error message using if else statement.
    
# These are some print statements or options to select an arithmetic operation. This check if the user input (choice) is one of the valid options or not.
def calculator():
    while True:
        print("Select operation:")
        print("1. Add")
        print("2. Subtract")
        print("3. Multiply")
        print("4. Divide")
        choice = input("Enter your choice or operation (1, 2, 3, 4,):  ")

# Now, We get two numbers from user to do calculations by variables and input functions.
        if choice in ['1','2','3','4']:
            num1 = float(input("Enter the first number: "))
            num2 = float(input("Enter the second number: "))
    
# After this, Now we use if/else statement to know what operation the user choose.
        if choice == '1':
            print(f"{num1} + {num2} = {add(num1, num2)}") 
        elif choice == '2':
            print(f"{num1} - {num2} = {subtract(num1, num2)}")
        elif choice == '3':
            print(f"{num1} * {num2} = {multiply(num1, num2)}")
        elif choice == '4':
            print(f"{num1} / {num2} = {divide(num1, num2)}")
        else:
            print("Invalid input") 
# We use one more else statement to handle invalid input.
    
        
calculator()
```
